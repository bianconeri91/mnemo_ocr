## 📤 Загрузка файла и OCR
Загрузите PNG/JPG или DOCX — система автоматически извлечёт изображения и выполнит OCR.

```python
uploaded = files.upload()

for fname, data in uploaded.items():
    print(f"\n📄 Обработка файла: {fname}")

    # Попытка декодировать как изображение
    arr = np.frombuffer(data, np.uint8)
    img = cv2.imdecode(arr, cv2.IMREAD_COLOR)

    if img is None:
        print("❌ Файл не является изображением. Если это DOCX, используйте локальный пайплайн.")
        continue

    # Пример — пользователь может заменить своими диапазонами
    color_ranges = {}

    title, sensors = extract_text(img, color_ranges)

    print("\n🏷 TITLE:", title)
    print("🧩 SENSORS:")
    for s in sensors:
        print(s)
```
